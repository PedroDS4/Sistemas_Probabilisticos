##**Atividade 2.1 de Sistemas Probabilísticos: Histograma 2D de duas variáveis**

Considere um braço robótico, com duas juntas rotacionais, fixo em um ponto no espaço, considerando que os parâmetros do braço são

$$
\begin{cases}
\vec{p}_1 = (\theta_1(t), l_1) \\
\vec{p}_2 = (\theta_2(t), l_2)
\end{cases}
$$

a posição da garra no espaço é dada então por

$$
\vec{r} = (x(\theta_1,\theta_2),y(\theta_1,\theta_2))
$$

onde

$$
\begin{cases}
x = l_1cos(\theta_1) + l_2cos(\theta_1 + \theta_2) \\
y = l_1sin(\theta_1) + l_2sin(\theta_1 + \theta_2)
\end{cases}
$$

sabendo que os ângulos seguem uma distribuição conjunta uniforme no plano cartesiano, temos então

$$
f_{\Theta_1,\Theta_2} (\theta_1,\theta_2) = \begin{cases}
\frac{1}{4\pi^2}, \ se \ -\pi \leq \theta_1, \theta_2 \leq \pi \\
0, \ c.c
\end{cases}
$$

como determinar então $f_{XY}(x,y)$?

Sabemos que o problema é modelado como

$$
\begin{cases}
x = g(\theta_1,\theta_2)\\
y = h(\theta_1,\theta_2)
\end{cases}
$$

o jacobiano é dado então por

$$
\mathbf{J}_{x,y}(\theta_1,\theta_2) =
\begin{bmatrix}
\frac{dx}{d\theta_1} & \frac{dx}{d\theta_2} \\
\frac{dy}{d\theta_1} & \frac{dy}{d\theta_2}
\end{bmatrix}
$$

calculando cada derivada

$$
\begin{cases}
\frac{dx}{d\theta_1} = -l_1sin(\theta_1) - l_2sin(\theta_1 + \theta_2) \\
\frac{dx}{d\theta_2} = - l_2sin(\theta_1 + \theta_2) \\
\frac{dy}{d\theta_1} = l_1 cos(\theta_1) + l_2cos(\theta_1 + \theta_2) \\
\frac{dy}{d\theta_2} = l_2cos(\theta_1 + \theta_2)
\end{cases}
$$

calculando o determinante do jacobiano, temos

$$
det(\mathbf{J}_{x,y}(\theta_1,\theta_2)) =  \frac{dx}{d\theta_1} \frac{dy}{d\theta_2} - \frac{dx}{d\theta_2} \frac{dy}{d\theta_1}
$$

substituindo as derivadas calculadas

$$
det(\mathbf{J}_{x,y}(\theta_1,\theta_2)) = (-l_1sin(\theta_1) - l_2sin(\theta_1 + \theta_2)) (l_2cos(\theta_1 + \theta_2)) - (-l_2sin(\theta_1 + \theta_2)) (l_1 cos(\theta_1) + l_2cos(\theta_1 + \theta_2))
$$

agora distribuindo os produtos das somas

$$
det(\mathbf{J}_{x,y}(\theta_1,\theta_2)) = -l_1sin(\theta_1)l_2cos(\theta_1 + \theta_2) - l_2^2sin(\theta_1 + \theta_2)cos(\theta_1 + \theta_2) + l_1l_2sin(\theta_1 + \theta_2)cos(\theta_1) + l_2^2sin(\theta_1 + \theta_2)cos(\theta_1 + \theta_2)
$$

cancelando os termos iguais com sinais opostos

$$
det(\mathbf{J}_{x,y}(\theta_1,\theta_2)) = -l_1sin(\theta_1)l_2cos(\theta_1 + \theta_2) + l_1l_2sin(\theta_1 + \theta_2)cos(\theta_1)
$$

ainda colocando $l_1l_2$ em evidência

$$
det(\mathbf{J}_{x,y}(\theta_1,\theta_2)) = l_1l_2(sin(\theta_1 + \theta_2)cos(\theta_1) - cos(\theta_1 + \theta_2)sin(\theta_1))
$$

Usando a identidade trigonométrica $sin(A-B) = sin(A)cos(B) - cos(A)sin(B)$, temos:

$$
det(\mathbf{J}_{x,y}(\theta_1,\theta_2)) = l_1l_2sin((\theta_1 + \theta_2) - \theta_1) = l_1l_2sin(\theta_2)
$$

Então precisamos apenas achar $sin(\theta_2)$ para calcular a distribuição conjunta almejada, e a fórmula final para $f_{XY}(x,y)$ será dada por

$$
f_{XY}(x,y) = \sum_{i} \frac{f_{\Theta_1 \Theta_2}(\theta_{1i},\theta_{2i})}{|det(\mathbf{J}(\theta_{1i},\theta_{2i}))|} = \frac{f_{\Theta_1 \Theta_2}(\theta^*_{1},\theta^*_{2})}{|det(\mathbf{J}(\theta^*_{1},\theta^*_{2}))|}
$$

onde $(\theta^*_{1}, \theta^*_{2})$ são as soluções para o sistema de equações $x = g(\theta^*_1, \theta^*_2)$ e $y = h(\theta^*_1, \theta^*_2)$, que podemos achar utilizando a definição das coordenadas x e y, onde podemos achar apenas o seno de $\theta_1$, assim:

$$
\begin{cases}
x = l_1cos(\theta_1) + l_2cos(\theta_1 + \theta_2) = l_1c_1 + l_2c_{12}\\
y = l_1sin(\theta_1) + l_2sin(\theta_1 + \theta_2) = l_1s_1 + l_2s_{12}
\end{cases}
$$

elevando ao quadrado e somando, temos

$$
x^2 + y^2 = l_1^2c_1^2 + l_1^2s_1^2 + 2l_1c_1l_2c_{12} +l_2^2c_{12}^2 + 2l_1s_1l_2s_{12}+l_2^2s_{12}^2 = l_1^2(c_1^2 +s_1^2) + 2l_1l_2(c_1c_{12} + s_1s_{12})
$$

desenvolvendo e usando as identidades trigonométricas

$$
\begin{cases}
   sin^2(\theta) + cos^2(\theta) = 1 \\
   cos(a+b - a) = cos(a)cos(a+b) + sin(a) sin(a+b) = cos(b)
\end{cases}
$$

$$
x^2 + y^2  = l_1^2+l_2^2 + 2l_1l_2(c_1c_{12} + s_1s_{12}) = l_1^2 + l_2^2 +2l_1l_2c_2
$$

isolando $c_2$

$$
c_2 = \frac{x^2 + y^2-l_1^2 - l_2^2}{ 2l_1l_2}
$$

elevando ao quadrado

agora usando $s_2^2 = 1 - c_2^2$, temos então

$$
s_2 = sin(\theta_2) = \pm \sqrt{1 - c_2^2} = \pm \sqrt{1 - (\frac{x^2 + y^2-l_1^2 - l_2^2}{ 2l_1l_2})^2}
$$

que são duas soluções possíveis para a equação da trajetória do robô.
Assim, finalmente podemos calcular a função densidade conjunta $f_{XY}(x,y)$ aplicando a fórmula mostrada acima, então

$$
f_{XY}(x,y) = \frac{\begin{cases} \frac{1}{4\pi^2 }, se \ -\pi \leq \theta_1, \theta_2 \leq \pi \\
0, \ \ c.c
\end{cases}}{|det(\mathbf{J}(\theta^*_{1},\theta^*_{2}))|}
$$


substituindo a solução do sistema de equações obitdo, finalmente temos

$$
f_{XY}(x,y) = 2 \cdot \begin{cases} \frac{1}{4\pi^2 l_1l_2 \sqrt{1 - (\frac{x^2 + y^2-l_1^2 - l_2^2}{ 2l_1l_2})^2}}, se \ -\pi \leq \theta_1, \theta_2 \leq \pi \\
0, \ \ c.c
\end{cases}
$$

O fator multiplicativo 2 se deve à equação ter um $\pm$ na sua solução, significando que há duas raízes, com módulos iguais porém sinais diferentes.

porém os limites dessa região ainda estão nas variáveis passadas, e temos ainda que o interior da raiz quadrada precisa ser maior, que zero, então analisando o domínio da função, temos ainda

$$
1 - (\frac{x^2 + y^2-l_1^2 - l_2^2}{ 2l_1l_2})^2 \geq 0
$$

então

$$
|\frac{x^2 + y^2-l_1^2 - l_2^2}{ 2l_1l_2}| \leq 1
$$

assim temos a seguinte desigualdade

$$
-1 \leq \frac{x^2 + y^2-l_1^2 - l_2^2}{ 2l_1l_2} \leq 1
$$

Resolvendo a desigualdade da direita

$$
x^2 + y^2 \leq l_1^2 + l_2^2 + 2l_1l_2 = (l_1+l_2)^2
$$

e agora desigualdade a da esquerda

$$
x^2 + y^2\geq  l_1^2 + l_2^2 - 2l_1l_2 \geq (l_1-l_2)^2
$$

ou seja, a região $XY$ de valores possíveis, está contida entre dois círculos, um de raio $(l_1-l_2)^2$, e um de $(l_1+l_2)^2$.

Então a densidade $f_{XY}(x,y)$ é finalmente dada por

$$
f_{XY}(x,y) = \begin{cases} \frac{1}{2\pi^2 l_1l_2 \sqrt{1 - (\frac{x^2 + y^2-l_1^2 - l_2^2}{ 2l_1l_2})^2}}, se \ (l_1-l_2)^2 \leq x^2 + y^2 \leq  (l_1+l_2)^2 \\
0, \ \ c.c
\end{cases}
$$

se $l_1 = l_2$, temos um caso especial onde as curvas de nível da densidade conjunta se tornam perfeitamente circulares com ciruclos de raio $l_1+l_2$ até o raio de $l_1-l_2$ com um pico de densidade nas bordas desses círculos, ou no centro, caso $l_1=l_2$.


###**Plot Scatter Duplo para visualizar a densidade**

```python
import numpy as np
import matplotlib.pyplot as plt




N = 10000
theta1 = np.random.uniform(-np.pi,np.pi,N)
theta2 = np.random.uniform(-np.pi,np.pi,N)

l_1 = 4
l_2 = 5

# x e y calculados como funções de \theta_1 e \theta_2
x = l_1*np.cos(theta1) + l_2*np.cos(theta1+theta2)
y = l_1*np.sin(theta1) + l_2*np.sin(theta1+theta2)



#Plot básico Scatter
plt.scatter(x,y, alpha = 0.1, color = 'blue')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Histograma de X e Y')
plt.show()
```

###**Histograma 3D para visualização por cima**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



# Histograma 3D
delta_x = (x.max() - x.min())/200
delta_y = (y.max() - y.min())/200
hist, xedges, yedges = np.histogram2d(x, y, bins=200, weights = np.ones_like(x)*(1/(delta_x*delta_y*len(x))))

# Centros dos Bins
xpos, ypos = np.meshgrid(xedges[:-1] + (xedges[1]-xedges[0])/2., yedges[:-1] + (yedges[1]-yedges[0])/2., indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

# Pontos em x e y
dx = dy = (xedges[1]-xedges[0]) * np.ones_like(zpos)
dz = hist.ravel()

# Colormap ponderado pelos pesos
colors = plt.cm.viridis(dz / dz.max())


# Plot 3D do Histograma
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average', color=colors)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Contagem')
ax.set_title('Histograma 3D Normalizado de X e Y')

# Ângulo de visualização(phi e azimute)
ax.view_init(elev=60, azim=45)
fig.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.viridis, norm=plt.Normalize(vmin=dz.min(), vmax=dz.max())), ax=ax, label='Contagem')
plt.show()
```

###**Visualização da PDF conjunta real**

```python
#Função Densidade Analítica
def f_XY(x,y):
  sqrt_term_arg = 1 - ((x**2 + y**2 - l_1**2 - l_2**2) / (2 * l_1 * l_2))**2
  density = 1 / (4 * np.pi**2 * l_1 * l_2 * np.sqrt(np.maximum(1e-4, sqrt_term_arg)))
  return density



# Cria o MeshGrid
x_vals = np.linspace(x.min(), x.max(), 1000) #
y_vals = np.linspace(y.min(), y.max(), 1000)
Xg, Yg = np.meshgrid(x_vals, y_vals)
Z = f_XY(Xg, Yg)


fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
surface_plot = ax.plot_surface(Xg, Yg, Z, cmap='viridis', edgecolor='none')
fig.colorbar(surface_plot, label='Densidade')
ax.set_title('Densidade conjunta f_XY(x,y) analítica')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Densidade')

# zoom no gráfico
ax.set_xlim([-(l_1+l_2), (l_1+l_2)])
ax.set_ylim([-(l_1+l_2), (l_1+l_2)])

# ax.set_zlim([0, 0.4])
ax.view_init(elev=60, azim=45)



plt.show()
```

Podemos ver que a densidade analítica se paroxima muito dos histogramas e gráficos feitos, sendo as coordenadas x e y tendo uma distribuição com picos nas bordas dos círculos que a delimitam.

##**Atividade 2.2 de Sistemas Probabilísticos: Teorema do Limite Central na prática**


Sejam $X_i$ com $i \in [0,n]$ variáveis aleatórias independentes de média zero e variância finita, considere agora a variável $Y$ dada pela combinação linear de $X_i$

$$
Y = \sum_{i = 1}^{n} a_i X_i
$$

considerando a combinação linear específica abaixo

$$
Y = \frac{1}{n} \sum_{i = 1}^{n} X_i
$$

a medida que $N$ se torna muito grande, a distribuição da variável $Y$ tende para uma distribuição gaussiana.



###**Média de Distribuições**
Para verificar o teorema do limite central, pegaremos 4 distribuições de probabilidade, especificamente a uniforme, gaussiana, laplaciana e exponencial de média nula, afim de verificar se a combinação linear

$$
\bar{X} = \sum_{i = 1}^{n} \frac{X_i}{\sqrt{n}}
$$

converge para uma distribuição gaussiana, para cada uma dessas distribuições.


Fazendo uma média de n variáveis aleatórias, incremetando esse $n$, ja é possível ver que a distribuição resultante se aproxima de uma normal cada vez mais, conforme aumentamos $n$.

Começando com uma média de apenas 2 termos, vemos que o padrão da gaussiana ja começa a aparecer

```python
import numpy as np
import matplotlib.pyplot as plt




#Número de termos da média(alto)
n = 2

#Número de amostras das distribuições
M = 10000


#Distribuição Uniforme
a = -3
b = 3

X_u_bar = np.zeros(M)

#Distribuição Exponencial
beta = 1
lambda_ = 1/beta
X_exp_bar = np.zeros(M)

X_exp_1 = np.random.exponential(beta,M)
X_exp_1 = X_exp_1 - lambda_


#Distribuição de Laplace
mu = 0
b = 1
X_laplace_bar = np.zeros(M)



#Distribuição Normal
mu_n = 0
sigma_n = 1

X_normal_bar = np.zeros(M)


# Loop para fazer a média
for i in range(n):
  Xu_1 = np.random.uniform(a,b,M)
  X_exp_1 = np.random.exponential(beta,M)- beta
  X_laplace_1 = np.random.laplace(loc = mu, scale = b, size = M)
  X_normal_1 = np.random.normal(mu_n, sigma_n, M)


  X_u_bar += Xu_1/np.sqrt(n)
  X_exp_bar += X_exp_1/np.sqrt(n)
  X_laplace_bar += X_laplace_1/np.sqrt(n)
  X_normal_bar += X_normal_1/np.sqrt(n)



# Plot dos Histogramas
bins = 100

np.histogram(X_u_bar, bins = bins)
np.histogram(X_exp_bar, bins = bins)
np.histogram(X_laplace_bar, bins = bins)
np.histogram(X_normal_bar, bins = bins)


plt.ylim(0,500)
plt.hist(X_u_bar, bins = bins, color = 'lime', edgecolor = "black", alpha = 0.7)
plt.hist(X_exp_bar, bins = bins, color = 'cyan', edgecolor = "black", alpha = 0.4)
plt.hist(X_laplace_bar, bins = bins, color = 'blue', edgecolor = "black", alpha = 0.2)
plt.hist(X_normal_bar, bins = bins, color = 'magenta', edgecolor = "black", alpha = 0.9)
plt.legend(['Uniforme', 'Exponencial', 'Laplace', 'Normal'])
plt.title(f'Histograma da Média das variáveis aleatórias com n = {n} termos')

plt.grid()
plt.show()
```

Aumentando o número de termos para 5, é possível ver que a tendência gaussiana começa a aumentar, como é mostrado na figura abaixo, e podemos aumentar ainda mais o numero para a distribuição final finalmente se tornar muito próxima de uma gaussiana.

```python
import numpy as np
import matplotlib.pyplot as plt




#Número de termos da média(alto)
n = 5

#Número de amostras das distribuições
M = 10000


#Distribuição Uniforme
a = -2
b = 2

X_u_bar = np.zeros(M)

#Distribuição Exponencial
beta = 1
lambda_ = 1/beta
X_exp_bar = np.zeros(M)

X_exp_1 = np.random.exponential(beta,M)
X_exp_1 = X_exp_1 - lambda_


#Distribuição de Laplace
mu = 0
b = 1
X_laplace_bar = np.zeros(M)



#Distribuição Normal
mu_n = 0
sigma_n = 1

X_normal_bar = np.zeros(M)


# Loop para fazer a média
for i in range(n):
  Xu_1 = np.random.uniform(a,b,M)
  X_exp_1 = np.random.exponential(beta,M)- beta
  X_laplace_1 = np.random.laplace(loc = mu, scale = b, size = M)
  X_normal_1 = np.random.normal(mu_n, sigma_n, M)


  X_u_bar += Xu_1/np.sqrt(n)
  X_exp_bar += X_exp_1/np.sqrt(n)
  X_laplace_bar += X_laplace_1/np.sqrt(n)
  X_normal_bar += X_normal_1/np.sqrt(n)



# Plot dos Histogramas
bins = 100

np.histogram(X_u_bar, bins = bins)
np.histogram(X_exp_bar, bins = bins)
np.histogram(X_laplace_bar, bins = bins)
np.histogram(X_normal_bar, bins = bins)


plt.ylim(0,500)
plt.hist(X_u_bar, bins = bins, color = 'lime', edgecolor = "black", alpha = 0.7)
plt.hist(X_exp_bar, bins = bins, color = 'cyan', edgecolor = "black", alpha = 0.4)
plt.hist(X_laplace_bar, bins = bins, color = 'blue', edgecolor = "black", alpha = 0.2)
plt.hist(X_normal_bar, bins = bins, color = 'magenta', edgecolor = "black", alpha = 0.9)
plt.legend(['Uniforme', 'Exponencial', 'Laplace', 'Normal'])
plt.title(f'Histograma da Média das variáveis aleatórias com n = {n} termos')
plt.grid()
plt.show()
```


##**Como fazer os histogramas finais ficarem com a mesma variância?**
Ja sabemos que com um número grande de variáveis aleatórias, a distribuição da variável $Y$ que é a média das variáveis aleatórias começa a se tornar gaussiana, podemos ainda, calculando as médias e variâncias das distribuições, garantir que a distribuição final tenha a mesma variãncia, pra cada uma.


###**Distribuição Uniforme**

A variância da distribuição uniforme pode ser calculada analiticamente usando sua densidade

$$
f_X(x) = \begin{cases} \frac{1}{b - a}, se \  \ a \leq x \leq b \\
0, \ \ c.c
\end{cases}
$$

calculando

$$
E \{\ (X-\mu)^2 \}\ = E\{\ X^2 \}\ - \mu^2
$$

onde $\mu = \frac{a+b}{2}$, e

$$
E\{\ X^2  \}\ = \int_{-\infty}^{\infty} x^2 f_X(x) dx = \frac{1}{b-a} \int_{a}^{b} x^2 dx = \frac{x^3}{3}|_{a}^{b} = \frac{1}{3} \frac{b^3 - a^3}{b-a} = (b-a)\frac{b^2 +ab + a^2}{3(b-a)} = \frac{b^2 + ab + a^2}{3}
$$

se $a = -b$

$$
E\{\ X^2  \}\ = \frac{b^2 -b^2 + b^2}{3} = \frac{b^2}{3}
$$

Podemos calcular a variância da distribuição exponencial, laplace da mesma maneira, e a variância da distribuição normal é o próprio desvio padrão da sua parametrização ao quadrado, assim temos


###**Distribuição Exponencial**

A densidade dessa distribuição com média zero é:

$$
f_X(x) = \begin{cases} \lambda e^{-\lambda ( x + \frac{1}{\lambda} ) }, se \ \ x \geq -\frac{1}{\lambda} \\
0, \ \ c.c
\end{cases}
$$

confirmando a média zero

$$
E\{\ X  \}\ = \int_{-\infty}^{\infty} x f_X(x) dx = \int_{-\frac{1}{\lambda}}^{\infty} x \lambda e^{-\lambda ( x + \frac{1}{\lambda} ) } dx = e^{-1} \int_{-\frac{1}{\lambda}}^{\infty} x \lambda e^{-\lambda  x  } dx
$$

utilizando integral por partes

$$
\begin{cases}
u = \lambda x \\
du = \lambda dx \\
dv = e^{-\lambda x }dx \\
v = \frac{-1}{\lambda}e^{-\lambda x}
\end{cases}
$$

temos então

$$
u(x) \cdot v(x)|_{-\frac{1}{\lambda}}^{\infty}  - \int_{-\frac{1}{\lambda}}^{\infty} v(x) du = \frac{-\lambda x}{\lambda}e^{-\lambda x}|_{-\frac{1}{\lambda}}^{\infty} - \int_{-\frac{\lambda}{\lambda}}^{\infty} \frac{-1}{\lambda}e^{-\lambda x} dx = \frac{e}{\lambda} - \frac{e}{\lambda} = 0
$$

então

$$
E\{\ X  \}\ = 0
$$

e agora a variãncia

$$
E\{\ X^2  \}\ = \int_{-\infty}^{\infty} x^2 f_X(x) dx = \int_{-\frac{1}{\lambda}}^{\infty} x^2 f_X(x) dx =  \int_{-\frac{1}{\lambda}}^{\infty} x^2 \lambda e^{-\lambda ( x + \frac{1}{\lambda} ) } dx = e^{-1} \int_{-\frac{1}{\lambda}}^{\infty} x^2 \lambda e^{-\lambda  x  } dx
$$

utilizando integral por partes mais uma vez, temos

$$
\begin{cases}
u = \lambda x^2 \\
du = 2x \lambda dx \\
dv = e^{-\lambda x }dx \\
v = \frac{-1}{\lambda}e^{-\lambda x}
\end{cases}
$$

substituindo na expressão

$$
u(x) \cdot v(x)|_{-\frac{1}{\lambda}}^{\infty}  - \int_{-\frac{1}{\lambda}}^{\infty} v(x) du = \frac{-\lambda x^2}{\lambda}e^{-\lambda x}|_{-\frac{1}{\lambda}}^{\infty} - \int_{-\frac{1}{\lambda}}^{\infty} \frac{-1}{\lambda}e^{-\lambda x} dx = \frac{e}{\lambda^2}
$$

finalmente temos

$$
E\{\ X^2  \}\ = \frac{e^{-1} \cdot e}{\lambda^2} = \frac{1}{\lambda^2}
$$


###**Distribuição de Laplace**

A densidade de Laplace é dada por
$$
f_X(x) = \frac{1}{2b} e^{\frac{-|x - \mu|}{b}}
$$

considerando ainda média zero

$$
f_X(x) = \frac{1}{2b} e^{\frac{-|x|}{b}}
$$

calculando sua variância


$$
E\{\ X^2  \}\ = \int_{-\infty}^{\infty} x^2 f_X(x) dx = \int_{-\infty}^{\infty} x^2 \frac{1}{2b} e^{\frac{-|x|}{b}} dx
$$

considerando ainda a paridade dessa distribuição

$$
E\{\ X^2  \}\ =  2\int_{0}^{\infty} x^2 \frac{1}{2b} e^{\frac{-x}{b}} dx
$$

aplicando integral por partes

$$
\begin{cases}
u = \frac{x^2}{2b} \\
du = \frac{x}{b} dx \\
dv = e^{\frac{-x}{b}}dx \\
v = -b e^{\frac{-x}{b}}
\end{cases}
$$

temos

$$
E\{\ X^2  \}\ =  2 ( u(x) \cdot v(x)|_{0}^{\infty}  - \int_{0}^{\infty} v(x) du) = 2 ( -\frac{x^2}{2b} b e^{\frac{-x}{2b}} |_{0}^{\infty}  - \int_{0}^{\infty} -b e^{\frac{-x}{b}} \frac{x}{b} dx  )
$$

sobrando então apenas

$$
E\{\ X^2  \}\ =  2 \int_{0}^{\infty} b e^{\frac{-x}{2b}} \frac{x}{b} dx  = 2 \int_{0}^{\infty} x e^{\frac{-x}{2b}}  dx  
$$

utilizando integração por partes mais uma vez


$$
\begin{cases}
u = x \\
du = dx \\
dv = e^{\frac{-x}{b}}dx \\
v = -b e^{\frac{-x}{b}}
\end{cases}
$$

temos

$$
\int_{0}^{\infty} x e^{\frac{-x}{2b}}  dx = -x b e^{\frac{-x}{b}} |_{0}^{\infty} - \int_{0}^{\infty} -b e^{\frac{-x}{b}} dx = b \int_{0}^{\infty} e^{\frac{-x}{b}} dx = -b^2 e^{\frac{-x}{b}} |_{0}^{\infty} = b^2
$$

então

$$
E\{\ X^2  \}\ = 2 b^2
$$


###**Igualando as variâncias**
As variâncias de cada uma das distribuições são

$$
\begin{cases}
VAR(X_u) = \frac{b^2}{3} \\
VAR(X_e) = \frac{1}{\lambda^2} \\
VAR(X_L) = 2c^2 \\
VAR(X_{𝒩}) = \sigma^2
\end{cases}
$$

para que todas a variância final igual, precisamos que todas essas variâncias sejam iguais, então temos

$$
\frac{b^2}{3} = \frac{1}{\lambda^2} = 2c^2 = \sigma^2
$$

isolando todos em função de $\sigma$, que é a variância da distribuição normal final, temos:

$$
\begin{cases}
b = \sqrt{3} \sigma \\
\lambda = \frac{1}{\sigma} \\
c = \frac{1}{\sqrt{2}} \sigma
\end{cases}
$$



###**Deixando os Histogramas Finais Iguais**
Agora, aplicando a equação obtida para os parâmetros da distribuição em função das variâncias obtidas acima, temos:



```python
import numpy as np
import matplotlib.pyplot as plt



#Número de termos da média(alto)
n = 100

#Número de amostras das distribuições
M = 10000

#Definindo o desvio padrão da distribuição final
sigma = 10


#Distribuição Uniforme
a = -np.sqrt(3)*sigma
b = np.sqrt(3)*sigma


X_u_bar = np.zeros(M)



#Distribuição Exponencial
lambda_ = 1/sigma
beta = 1/lambda_
X_exp_bar = np.zeros(M)




#Distribuição de Laplace
mu = 0
c = sigma/np.sqrt(2)


X_laplace_bar = np.zeros(M)



#Distribuição Normal
mu_n = 0
sigma_n = sigma

X_normal_bar = np.zeros(M)


# Loop para fazer a média
for i in range(n):

  Xu_1 = np.random.uniform(a,b,M)
  X_exp_1 = np.random.exponential(beta,M)- beta
  X_laplace_1 = np.random.laplace(loc = mu, scale = c, size = M)
  X_normal_1 = np.random.normal(mu_n, sigma_n, M)


  X_u_bar += Xu_1/np.sqrt(n)
  X_exp_bar += X_exp_1/np.sqrt(n)
  X_laplace_bar += X_laplace_1/np.sqrt(n)
  X_normal_bar += X_normal_1/np.sqrt(n)



# Plot dos Histogramas
bins = 100

np.histogram(X_u_bar, bins = bins)
np.histogram(X_exp_bar, bins = bins)
np.histogram(X_laplace_bar, bins = bins)
np.histogram(X_normal_bar, bins = bins)


plt.ylim(0,500)
plt.hist(X_u_bar, bins = bins, color = 'lime', edgecolor = "black", alpha = 0.8)
plt.hist(X_exp_bar, bins = bins, color = 'cyan', edgecolor = "black", alpha = 0.6)
plt.hist(X_laplace_bar, bins = bins, color = 'black', edgecolor = "black", alpha = 0.1)
plt.hist(X_normal_bar, bins = bins, color = 'magenta', edgecolor = "black", alpha = 0.4)
plt.legend(['Uniforme', 'Exponencial', 'Laplace', 'Normal'])
plt.title(fr'Histograma da Média das variáveis aleatórias com n = {n} distribuições e variância final $\sigma$ = {sigma}')
plt.grid()
plt.show()
```
